# Cross Validation of Model

#### in-sample training is using the data to test how accurate our model is however we do not know how does it fit with the real world data
#### Measuring Generaliztion performance involves testing the model with the data that has not been used to train it

### Importing the required dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../automobiles_fuel_dummies.csv")
```

```python
df.head(10)
```

### Splitting data into train and test subsets

```python
x_data = df[['highway-mpg']]
y_data = df[['price']]

x_train,x_test,y_train,y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state = 0)
```

### Cross validation

#### Randomly shuffling the test and train data set from the given sample and calculating the generalizing error

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
scores = cross_val_score(lr,x_data,y_data,cv=3)
```

```python
scores
```

### The above array are the R squared values as a result of testing and training with 3 sets of random test and train data

#### Following function shows the predicted values too

```python
from sklearn.model_selection import cross_val_predict
```

```python
values = cross_val_predict(lr,x_data,y_data,cv=3)
```
