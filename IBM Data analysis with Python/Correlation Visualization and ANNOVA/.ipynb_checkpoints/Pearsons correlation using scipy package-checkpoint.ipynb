# Pearsons correlation

```python
import pandas as pd
import scipy.stats as sp
```

```python
df = pd.read_csv("../automobiles_fuel_dummies.csv")
df = df.drop(df.columns[[0]],axis =1)
```

```python
df['horsepower'] = df['horsepower'].replace('?',0)
df['horsepower'] = df['horsepower'].astype('float64')
df['horsepower'] = df['horsepower'].replace(0,df['horsepower'].mean())
```

```python
df['horsepower']
```

### Pearsons correlation below gives (correlation coeff , p-value)

```python
sp.pearsonr(df['horsepower'],df['price'])
```

### Strong correlation with significant p-value

### Let us confirm by plotting the scatter plot

```python
import seaborn as sb
```

```python
sb.regplot(x = 'horsepower' , y ='price', data = df)
```

### Therefore we see that there is a strong correlation between price and horsepower

### Correlation between all the variables in the df

```python
df.corr()
```
