### MATPLOTLIB SCATTER PLOT

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=df["engine-size"],y=df["price"])
plt.title("Scatter plot for Engine Size vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Price")
```
