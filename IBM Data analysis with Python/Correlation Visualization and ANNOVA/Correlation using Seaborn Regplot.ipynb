### Correlation between variables in the data frame

```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
df = pd.read_csv("../automobiles_fuel_dummies.csv")
df = df.drop(df.columns[[0]],axis=1)
```

### Plotting the correlation between engine-size and price

```python
sb.regplot(x = 'engine-size', y = 'price', data = df)
```

### We see a positive linear correlation between engine-size and price therefore engine size is a good predictor of price

### Correlation between highway-mpg and price

```python
sb.regplot(x = 'highway-mpg' , y = 'price',data=df)
```

### Negative correlation between highway-mpg and price

```python
df['peak-rpm'] = df['peak-rpm'].replace('?',0)
```

```python
df['peak-rpm'] = df['peak-rpm'].astype('int64')
```

```python
df['peak-rpm'] = df['peak-rpm'].replace(0,df['peak-rpm'].mean())
```

```python
df['peak-rpm']
```

### Correlation between peak rpm and price

```python
sb.regplot(x = 'peak-rpm',y='price',data=df)
```

### Weak correlation is observed between the price and the peak-rpm therefore peak-rpm is cannot be used as a predictor of price
