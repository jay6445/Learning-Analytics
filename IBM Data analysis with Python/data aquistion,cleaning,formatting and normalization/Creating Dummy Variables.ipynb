# Converting categorical variables to dummy variables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../automobile_price_binned.csv")
```

```python
df = df.drop(df.columns[[0]],axis=1)

df.to_csv("automobile_price_binned.csv")
```

```python
df.head(20)
```

## FUEL is of two types GAS or DIESEL

### Create dummy variables for fuel

```python
fuel_dummies = pd.get_dummies(df["fuel-type"])

fuel_dummies
```

### Adding this data frame to the source df

```python
df = pd.concat([df, fuel_dummies], axis=1, join='inner')
```

```python
df
```

```python
df.to_csv("automobiles_fuel_dummies.csv")
```
