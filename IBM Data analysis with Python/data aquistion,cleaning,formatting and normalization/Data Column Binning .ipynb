# Binning is used to divide the data in different category that helps us understand the column better

## Like the prices of cars in our case are divided into categories low medium and high

```python
import pandas as pd
import numpy as np
```

```python
# We will load the data from our formatted dataset

df = pd.read_csv("../automobile_formatted.csv")
```

```python
df
```

```python
df = df.drop(df.columns[[0]],axis =1)
```

# write the updated data to csv

df.to_csv("automobile_formatted.csv")

### Create Bins

```python
bins = np.linspace(df["price"].min(),df["price"].max(),4)
```

```python
bins
```

### Create Group names

```python
group_names = ["low","medium","high"]
```

### create new column and assign group names to bins

```python
df["price-binned"] = pd.cut(df["price"],bins,labels = group_names)
```

```python
df.head(30)
```

### Write binned column data to csv

```python
df.to_csv("automobile_price_binned.csv")
```
