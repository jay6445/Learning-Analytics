# Formatting the datatypes

import pandas as pd

import numpy as np

```python
df = pd.read_csv("../imports-85.data",header = None)
```

```python
## adding headers

headers=["symboling",
"normalized-losses",        
"make",                                          
"fuel-type",        
"aspiration",             
"num-of-doors",           
"body-style",              
"drive-wheels",         
"engine-location",        
"wheel-base",              
"length",               
"width",                  
"height",                
"curb-weight",           
"engine-type",           
"num-of-cylinders",     
"engine-size",            
"fuel-system",             
"bore",                 
"stroke",                 
"compression-ratio",      
"horsepower",              
"peak-rpm",             
"city-mpg",             
"highway-mpg",             
"price"]

df.columns = headers 
```

```python
df.head(10)
```

## Checking the datatypes of the columns

df.dtypes

## cleaning the dataset

df.dropna(subset=["price"],axis=0,inplace=True)

```python
df.head(10)
```

## replacing string '?' with np.nan

df["price"] = df["price"].replace('?',np.nan)

```python
df.head(10)
```

## dropping the nan columns 

df.dropna(subset = ["price"],axis=0, inplace = True)

```python
df.head(10)
```

## formatting the price column

#converting object to int

df["price"].astype("int")

```python
df.head(10)
```

## cleaning the normalized-losses column

df["normalized-losses"]


## we will first replace the '?' string with np.nan

df["normalized-losses"] = df["normalized-losses"].replace(np.nan,0.0)

```python
df.head(10)
```

```python
df.dtypes

```

## converting normalized-losses from object to numeric

df["normalized-losses"] = df["normalized-losses"].astype('int64')

```python
df.dtypes
```

## Now replacing the zero values with the mean value of the column

mean = df["normalized-losses"].mean()

df["normalized-losses"] = df["normalized-losses"].replace(0,mean)

```python
df.head(60)
```

## lets further format the column upto 2 decimal places

df["normalized-losses"] = df["normalized-losses"].round(decimals=2)

```python
df.head(50)
```

# We formatted the data columns from object to integer for easy analysis




```python
# Exporting our cleaned data set to csv

df.to_csv("automobile_formatted.csv")
```
