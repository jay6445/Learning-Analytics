
# Data Wrangling/Cleaning/Preprocessing

```python
import pandas as pd

import numpy as np
```

```python
df = pd.read_csv("../imports-85.data",header = None)
```

```python
df
```

```python
# adding headers

headers=["symboling",
"normalized-losses",        
"make",                                          
"fuel-type",        
"aspiration",             
"num-of-doors",           
"body-style",              
"drive-wheels",         
"engine-location",        
"wheel-base",              
"length",               
"width",                  
"height",                
"curb-weight",           
"engine-type",           
"num-of-cylinders",     
"engine-size",            
"fuel-system",             
"bore",                 
"stroke",                 
"compression-ratio",      
"horsepower",              
"peak-rpm",             
"city-mpg",             
"highway-mpg",             
"price"]

df.columns = headers 
```

```python
df.head(10)
```

# removing/dropping the entries with missing price

df.dropna(subset=["price"], axis = 0) #does not modify the original dataframe


```python
df.dropna(subset=["price"], axis = 0, inplace = True) #modifies the dataframe
```

```python
df.head(10)
```

# Replacing the missing values

# we will replace the missing values in the normalized losses column with the column average

#first converting the datatype from object to float64
df["normalized-losses"] = pd.to_numeric(df["normalized-losses"],errors="coerce")

df.head(200)

# calculating the mean

mean = df["normalized-losses"].mean()

mean 

# now replacing the missing values with mean value 

df["normalized-losses"] = df["normalized-losses"].replace(np.nan,mean)
